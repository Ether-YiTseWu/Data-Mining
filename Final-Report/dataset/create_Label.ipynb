```python
import pandas as pd
import numpy as np
import os
```

```python
img_list = []
label_list = []

# append list
for filename in os.listdir('./'):
    if filename.endswith('.jpg'):
        img_list.append(filename)
        label = int(filename[-5]) - 1
        if label == -1:
            label = 9
        label_list.append(label)
```

```python
img_df = pd.DataFrame(img_list, columns = ['image'])
label_list = pd.DataFrame(label_list, columns = ['label'])
df = pd.concat([img_df, label_list], axis = 1)
df
```

```python
df.to_csv('label.csv')
```
