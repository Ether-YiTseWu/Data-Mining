```python
import numpy as np 
import pandas as pd
```

```python
df = pd.read_csv("./chinese_mnist.csv")
df_0 = df[df['character']=="零"]
df_1 = df[df['character']=="一"]
df_2 = df[df['character']=="二"]
df_3 = df[df['character']=="三"]
df_4 = df[df['character']=="四"]
df_5 = df[df['character']=="五"]
df_6 = df[df['character']=="六"]
df_7 = df[df['character']=="七"]
df_8 = df[df['character']=="八"]
df_9 = df[df['character']=="九"]
df_all = [df_0, df_1, df_2, df_3, df_4, df_5, df_6, df_7, df_8, df_9]

df_0to9 = pd.concat([df_0, df_1], axis = 0)
for i in range(2,10):
    df_0to9 = pd.concat([df_0to9, df_all[i]], axis = 0)
df_0to9
```

```python
df_label = df_0to9.rename(columns={'character':'label'})
df_label = df_label.reset_index()
df_label = df_label[['label']]
df_label
```

```python
image = []
for i in range(10000):
    name = 'input_' + str(df_0to9.iloc[i]['suite_id']) + '_' + str(df_0to9.iloc[i]['sample_id']) + '_' + str(df_0to9.iloc[i]['code']) + '.jpg'
    image.append(name)
image
```

```python
df_image = pd.DataFrame(image)
df_image = df_image.rename(columns={0:'image'})
df_image
```

```python
df_final = pd.concat([df_image, df_label], axis = 1)
df_final
```

```python
imageFile = []
for i in range(10000):
    imageFile.append(df_final.iloc[i]['image'])
```

```python
import os
# del no need img
for filename in os.listdir('./'):
    if filename.endswith('.jpg') and filename not in imageFile:
        print(filename)
        os.remove(filename)
```
